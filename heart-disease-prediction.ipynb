```python
#pca exercise
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/18_PCA/Exercise/heart.csv')
df.describe()
import matplotlib.pyplot as plt
pd.crosstab(df.ChestPainType,df.Sex).plot(kind='bar')
pd.crosstab(df.Sex,df.HeartDisease).plot(kind = 'bar')
pd.crosstab(df.Age,df.Sex).plot(kind = 'bar')
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
scaler = MinMaxScaler()
df = pd.get_dummies(df,drop_first = True)
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
x = df.drop('HeartDisease',axis = 'columns')
y = df.HeartDisease
scores=[]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)




scores

from sklearn.model_selection import cross_val_score
import numpy as np

```

```python
# outliers
df.head()
df[df.Cholesterol>(df.Cholesterol.mean()+3*df.Cholesterol.std())]
```

```python
model = [SVC,RandomForestClassifier,LogisticRegression]
l = SVC()
rf = RandomForestClassifier()
lr =LogisticRegression(max_iter=100)

```

```python
#logistic regression
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
# svm
l.fit(x_train,y_train)
l.score(x_test,y_test)
```

```python
# random forest
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
# cross validation to average their accuracy
z = []
z.append(np.average(cross_val_score(SVC(),x_train,y_train)))
z.append(np.average(cross_val_score(RandomForestClassifier(),x_train,y_train)))
z.append(np.average(cross_val_score(LogisticRegression(max_iter=100),x_train,y_train)))
z
```

```python
#pca


from sklearn.decomposition import PCA
pca = PCA(0.95)# retain 95% useful features or information or variation and create new dimensions
x_pca = pca.fit_transform(x)


```

```python
# cross validation to average their accuracy
x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size = 0.2)
z = []
z.append(np.average(cross_val_score(SVC(),x_train,y_train)))
z.append(np.average(cross_val_score(RandomForestClassifier(),x_train,y_train)))
z.append(np.average(cross_val_score(LogisticRegression(max_iter=100),x_train,y_train)))
z
```
